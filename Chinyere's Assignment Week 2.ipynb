# Write a program with the following two parts:
### Part A: Create a list of at least 5 items (for example, a shopping list, fruit names, or country names). Use a for loop to print each item in the list along with its position number, starting from 1. The output should look like: 1. Apple, 2. Mango, and so on.
### Part B: Write a while loop that keeps asking the user to enter a password. If the password entered matches a password you have set in your code, print Access granted and stop the loop. If it does not match, print Wrong password, try again and let the user try again. After 3 failed attempts, print Too many attempts. Access denied. and exit the loop using break.


```python
# Part A: List with for loop

items = ["Apple", "Mango", "Banana", "Orange", "Grapes"]

print("List of items:")
for a in range(len(items)):
    print(f"{a + 1}. {items[a]}")
```

```python
# Part B: Password with while loop

correct_password = "chy123"
attempts = 0
highest_attempts = 3

while True:
    user_input = input("Enter password: ")
    
    if user_input == correct_password:
        print("Access granted")
        break
    else:
        attempts += 1
        if attempts >= highest_attempts:
            print("Too many attempts. Access denied.")
            break
        else:
            print("Wrong password, try again")
```

# Define and call the following functions:
### greet(name): takes a name as a positional argument and prints a welcome message such as: Hello [name], welcome to Python class.
### calculate_bmi(weight, height=1.75): uses a keyword argument with a default value. The function should return and print the BMI result calculated as weight divided by height squared. Test the function twice, one providing both arguments and one with only the weight.
### total_score(*args): accepts any number of test scores. The function should return and print the total sum and the average of all scores passed in. Test it with at least three different calls using a different number of arguments each time.



```python
# Function 1: greet
def greet(name):
    print(f"Hello {name}, welcome to Python class.")

# Call the function
greet("Chinyere")
```

```python
# Function 2: calculate_bmi
def calculate_bmi(weight, height=1.75):
    bmi = weight / (height ** 2)
    print(f"BMI: {bmi:.2f}")
    return bmi
```

```python
# Test the function
calculate_bmi(50, 1.70)   # both arguments
calculate_bmi(35)         # only weight (uses default height)
```

```python
# Function 3: total_score
def total_score(*args):
    total = sum(args)
    average = total / len(args)
    print(f"Total: {total}")
    print(f"Average: {average:.2f}")
    return total, average
```

```python
# Test the function with different numbers of arguments
total_score(80, 90, 85)
print()

total_score(70, 75, 80, 85)
print()

total_score(60, 65)
```

# Write a program that does the following:

```python
# Step 1: Write to file (write mode)
with open("notes.txt", "w") as file:
    file.write("This is alien attack.\n")
    file.write("You are the first target.\n")
    file.write("Give your last wish.\n")
```

# Write a program that does the following:
### Create a CSV file called products.csv with the columns Product, Price and Quantity. Add at least 4 rows of product data to the file using csv.writer().
### Read the file back using csv.DictReader() and print each row to the screen.


```python
import csv
```

```python
# Step 1: Write to CSV file
with open("products.csv", "w", newline="") as file:
    writer = csv.writer(file)

     # Write header
    writer.writerow(["Product", "Price", "Quantity"])
    
    # Write rows
    writer.writerow(["Gold", 1.2, 10])
    writer.writerow(["LNG", 0.5, 20])
    writer.writerow(["Nuclear weapon", 0.8, 15])
    writer.writerow(["Robots", 1.5, 8])
```

```python
# Read from CSV file
with open("products.csv", "r") as file:
    reader = csv.DictReader(file)
    
    for row in reader:
        print(row)
```

# Part B: Pandas

```python
import pandas as pd
```

```python
# Create DataFrame
data = {
    "Name": ["Alice", "Bob", "Charlie", "David", "Eva"],
    "Age": [20, 21, 19, 22, 20],
    "Score": [55, 40, 65, 70, 45],
    "Grade": ["B", "F", "A", "A", "C"]
}

df = pd.DataFrame(data)

# Print first 3 rows
print("\nFirst 3 rows:")
print(df.head(3))
```

```python
# Filter students with Score > 50
print("\nStudents with Score > 50:")
print(df[df["Score"] > 50])

# Sort by Score descending
print("\nSorted by Score (descending):")
df_sorted = df.sort_values(by="Score", ascending=False)
print(df_sorted)

# Add Passed column using np.where()
df['Passed'] = np.where(df['Score'] >= 50, True, False)

print("\nDataFrame with Passed column:")
print(df)
```

```python
# Save to CSV
df.to_csv("class_results.csv", index=False)
```
